Merge together all rows sharing the same book title, author and publisher. We will call the resulting datset merged books. The books that have not been merged together will not appear in merged books

```python
merge = books.groupby(['Publisher','Book-Title','Book-Author'], as_index = False).count()
merged_books = merge[merge['ISBN']>1]
merged_books.head()
```

For each book in merged books compute its average rating.

```python
avg_rat = pd.merge(merged_books, avg, on = ['Book-Title', 'Book-Author', 'Publisher'])
avg_rat.head()
```

```python
average_rating = avg_rat.groupby(['Book-Title', 'Book-Author', 'Publisher'], as_index= False)['Book-Rating'].mean()
average_rating.head()
```

Assuming that the number of reviews with a text (column work_text_reviews_count) is split equally among all authors, find for each authors the total number of reviews with a text. We will call this quantity the shared number of reviews with a text.

```python
goodbooks = pd.read_csv('goodbooks.csv' , encoding = 'latin-1')
goodbooks_ratings = pd.read_csv('goodbooks-Ratings.csv' , encoding = 'latin-1')
```

```python
goodbooks[['original_title', 'authors', 'work_text_reviews_count']].head(20).sort_values('original_title')
```

```python
sum_reviews = goodbooks['work_text_reviews_count'].sum()
sum_authors = goodbooks['authors'].count()
shared_number_of_reviews_with_a_text_partial = sum_reviews / sum_authors 
```

```python
qw = goodbooks.groupby('authors', as_index=False).count()[['authors', 'book_id']]
qw['shared_number_of_reviews_with_a_text_partial'] = (sum_reviews/sum_authors)
qw['shared_number_of_reviews_with_a_text'] = (shared_number_of_reviews_with_a_text_partial * qw['book_id'])
qw
```

```python
goodbooks_new = pd.merge(goodbooks, qw, on='authors')
```
